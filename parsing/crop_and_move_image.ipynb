Создает папку для картинки в датасете и сразу же обрезает ее.

```python
import os
import shutil
from PIL import Image
```

```python
source = "images_test"
```

```python
for file in os.listdir(source):
    index = file.partition('_')[0]
    print(index)
    print(f"{source}/{file}")
    os.makedirs(f"dataset/{index}", exist_ok=True)
    shutil.move(f"{source}/{file}", f"dataset/{index}")
    img = Image.open(f"dataset/{index}/{file}")
    res = img.crop((37, 220, 450, 630))
    res.save(f"dataset/{index}/{file}_1.jpg")
    res = img.crop((37, 672, 450, 1082))
    res.save(f"dataset/{index}/{file}_2.jpg")
    res = img.crop((496, 222, 1032, 630))
    res.save(f"dataset/{index}/{file}_3.jpg")
    res = img.crop((496, 672, 1032, 1082))
    res.save(f"dataset/{index}/{file}_4.jpg")
    res = img.crop((1340, 260, 1582, 505))
    res.save(f"dataset/{index}/{file}_5.jpg")
#x = "9_3D-Macula-6x6_RETINA01_Portrait_001"
#x.partition('_')[0]
```
