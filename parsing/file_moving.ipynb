Скрипт чтобы переместить картинки и xml из папок датасета в отдельные папки для картинок и xml. 

1. Создается папка files
2. В files копируются папки датасет у которых имя фамилия или число (например можно скопировать "Агафонова", "Александрова" и т.д)
3. Потом запускаем скрипты и автоматом будут извлечены картинки и xml файлы

```python
import os
import shutil
from pathlib import Path
import operator as op
```

```python
source = Path(r"files")
destination_img = "images"
destination_xml = "xml_files"
```

```python
for root, dirs, files in os.walk(source):
    for name in files:
        if (name.endswith((".jpg")) and op.contains(name, "Macula")==1):
            #print(root)
            #print(name)
            shutil.move(f"{root}/{name}", destination_img)
```

```python
for root, dirs, files in os.walk(source):
    for name in files:
        if (name.endswith((".xml")) and op.contains(name, "_Information")==0):
            #print(root)
            #print(name)
            shutil.move(f"{root}/{name}", destination_xml)
```
