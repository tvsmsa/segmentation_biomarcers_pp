```python
import pandas as pd
```

```python
df = pd.read_csv("all_data.csv")
```

```python
df.head()
```

```python
df_retina = df[df['analysis_module']=='RETINA']
df_retina = df_retina.drop(["cd_area_ratio", "cup_area", "cup_volume", "disc_area", "horizontal_dd", "linear_cdr", "rim_area", "rim_volume", "vertical_cdr", "vertical_dd", "analysis_module"], axis=1)
df_retina = df_retina[~df['filename'].str.contains('-', na=False)]
df_retina.reset_index(inplace=True)
df_retina.index+=1
df_retina
```

```python
df_retina.to_csv("retina_data.csv")
```

```python
df_nfl = df[df['analysis_module']=='NFL']
#Столбцы для RETINA
df_nfl = df_nfl.drop(["average_thickness", "center_thickness", "etdrs_inT", "etdrs_outI", "etdrs_outN", "etdrs_outS", "etdrs_outT", "etdrs_center", "etdrs_inI", "etdrs_inN", "etdrs_inS", "total_volume"], axis=1)
df_nfl = df_nfl.drop("report_title", axis=1)
df_nfl
```

```python
df_nfl.to_csv("nfl_data.csv")
```
